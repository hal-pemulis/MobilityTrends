```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '../Resources/applemobilitytrends-2020-04-13.csv'

mobility_df = pd.DataFrame(pd.read_csv(path))

mobility_df.head()
```

```python
mobility_df["region"].unique()
```

```python
usa_mobility = mobility_df.loc[mobility_df['region']=='United States']
usa_driving = usa_mobility.loc[usa_mobility['transportation_type']=='driving'].set_index('region')
usa_driving = usa_driving.drop(columns=['geo_type', 'transportation_type']).T

usa_driving 
```

```python
usa_walking = usa_mobility.loc[usa_mobility['transportation_type']=='walking'].set_index('region')
usa_walking = usa_walking.drop(columns=['geo_type', 'transportation_type']).T
usa_walking
```

```python
usa_line = usa_driving.plot.line(figsize=(15,10), title='2020 Apple Maps routing requests (US)', legend=False)
usa_line.set_ylabel('Apple Maps routing requests')
plt.axvline(56, color='orange', linestyle='dotted')
```

```python
germany_mobility = mobility_df.loc[mobility_df['region']=='Germany']
german_driving = germany_mobility.loc[germany_mobility['transportation_type']=='driving'].set_index('region')
german_driving = german_driving.drop(columns=['geo_type', 'transportation_type']).T

italy_mobility = mobility_df.loc[mobility_df['region']=='Italy']
italy_driving = italy_mobility.loc[italy_mobility['transportation_type']=='driving'].set_index('region')
italy_driving = italy_driving.drop(columns=['geo_type', 'transportation_type']).T
```

```python
germany_line = german_driving.plot.line(figsize=(15,10), title='2020 Apple Maps routing requests (Germany)', legend=False)
germany_line.set_ylabel('Apple Maps routing requests')
plt.axvline(56, color='orange', linestyle='dotted')
```
